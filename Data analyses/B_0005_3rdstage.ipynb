```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
file_path = '/Users/kritikasethi10/Desktop/ml-project/ml-semper-fortis/Datasets/Cleaned Stage 3/B0005_no_outliers.csv'
df = pd.read_csv(file_path)
```

```python
# Shape and data types
print(f"Shape of the dataset: {df.shape}")
print(df.dtypes)
```

```python
df.head()
```

```python
# Summary statistics
print(df.describe())
```

```python
columns = ['Voltage_measured', 'Current_measured', 'Temperature_measured', 'Capacity']

for column in columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# before smoothing
plt.figure(figsize=(14, 8))
plt.plot(df['Cycle_Number'], df['Capacity'], label='Capacity')
plt.title('Capacity vs. Cycle Number')
plt.xlabel('Cycle Number')
plt.ylabel('Capacity (Ahr)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Voltage_measured'])
plt.xlabel('Time')
plt.ylabel('Voltage (V)')
plt.title('Voltage vs. Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Current_measured'])
plt.xlabel('Time')
plt.ylabel('Current (A)')
plt.title('Current vs. Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], df['Temperature_measured'], label='Battery Temperature')
plt.plot(df['Time'], df['Ambient_temperature'], label='Ambient Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature vs. Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Voltage_measured'], df['Current_measured'])
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.title('Current vs. Voltage')
plt.show()
```
